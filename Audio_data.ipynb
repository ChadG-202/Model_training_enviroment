```python
# Imports
import librosa, librosa.display
import matplotlib.pyplot as plt
import numpy as np
import IPython.display as ipd
```

```python
# Variables
happy = "Data/Train/Pre/Audio/Happy/0.wav"
neutral = "Data/Train/Pre/Audio/Neutral/100.wav"
sad = "Data/Train/Pre/Audio/Sad/200.wav"

# audio data, smaple rate
happy_signal, sr = librosa.load(happy, sr=22050)
neutral_signal, sr = librosa.load(neutral, sr=22050)
sad_signal, sr = librosa.load(sad, sr=22050)

# Load
ipd.Audio(happy)
```

```python
ipd.Audio(neutral)
```

```python
ipd.Audio(sad)
```

```python
# Display amplitude graph
# Happy
librosa.display.waveplot(happy_signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
# Neutral
librosa.display.waveplot(neutral_signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
# Sad
librosa.display.waveplot(sad_signal, sr=sr)
plt.xlabel("Time")
plt.ylabel("Amplitude")
```

```python
# Display magnitude/frequency graph
# Happy
fft = np.fft.fft(happy_signal) # Fast Fourier Transform

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]
                           
plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```

```python
# Neutral
fft = np.fft.fft(neutral_signal)

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]
                           
plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```

```python
# Sad
fft = np.fft.fft(sad_signal)

magnitude = np.abs(fft)
frequency = np.linspace(0, sr, len(magnitude))

left_frequency = frequency[:int(len(frequency)/2)]
left_magnitude = magnitude[:int(len(frequency)/2)]
                           
plt.plot(left_frequency, left_magnitude)
plt.xlabel("Frequency")
plt.ylabel("Magnitude")
```

```python
# Frequency against time graph
# Happy
n_fft = 2048
hop_length = 512

stft = librosa.core.stft(happy_signal, hop_length=hop_length, n_fft=n_fft)

spectrogram = np.abs(stft)

log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
```

```python
# Neutral
n_fft = 2048
hop_length = 512

stft = librosa.core.stft(neutral_signal, hop_length=hop_length, n_fft=n_fft)

spectrogram = np.abs(stft)

log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
```

```python
# Sad
n_fft = 2048
hop_length = 512

stft = librosa.core.stft(sad_signal, hop_length=hop_length, n_fft=n_fft)

spectrogram = np.abs(stft)

log_spectrogram = librosa.amplitude_to_db(spectrogram)

librosa.display.specshow(log_spectrogram, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("Frequency")
plt.colorbar()
```

```python
# Mel-frequency cepstral coefficients over time graph
# Happy
MFFCs = librosa.feature.mfcc(happy_signal, n_fft=n_fft, hop_length=hop_length, n_mfcc=13)

librosa.display.specshow(MFFCs, sr=sr, hop_length=hop_length)
plt.xlabel("Time")
plt.ylabel("MFCC")
plt.colorbar()
```
